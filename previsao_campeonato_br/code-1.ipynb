```python
import pandas as pd
import requests

requisicao = requests.get("https://pt.wikipedia.org/wiki/Campeonato_Brasileiro_de_Futebol_de_2023_-_S%C3%A9rie_A")

tabelas = pd.read_html(requisicao.text)

# for tabela in tabelas:
#    display(tabela) #visualizar todas as tabelas da página

tabela_classificacao = tabelas[6]
tabela_jogos = tabelas[7]
display(tabela_classificacao)
display(tabela_jogos)
```

```python
nomes_times = list(tabela_jogos["Casa \ Fora"])
nomes_abreviados = list(tabela_jogos.columns)
nomes_abreviados.pop(0)

# criando dicionário para mesclar os nomes com as abreviações

de_abv_para_nomes = dict(zip(nomes_abreviados, nomes_times)) 
print(de_abv_para_nomes)
```

Formatando tabela de acordo com os nomes

```python
# define a coluna "Casa \ Fora" como o índice do DataFrame tabela_jogos
tabela_jogos_formatada = tabela_jogos.set_index("Casa \ Fora") # a coluna não será mais tratada como uma coluna regular mas sim como o índice das linhas do DataFrame

tabela_jogos_formatada = tabela_jogos_formatada.unstack().reset_index() # reorganiza a visualização das colunas
tabela_jogos_formatada = tabela_jogos_formatada.rename(columns={"level_0": "Fora", "Casa \ Fora": "Casa", 0:"Resultado"}) # troca o nome das colunas

# função para trocar as abreviações pelos nomes dos times
def ajustar_abv_time(Linha):
    abv = Linha["Fora"]
    nome = de_abv_para_nomes[abv]
    return nome

tabela_jogos_formatada["Fora"] = tabela_jogos_formatada.apply(ajustar_abv_time, axis=1)

# pega as linhas onde a coluna Fora for diferente da coluna Casa
tabela_jogos_formatada = tabela_jogos_formatada[tabela_jogos_formatada["Fora"]!=tabela_jogos_formatada["Casa"]]
display(tabela_jogos_formatada)
```

```python
# separando as tabelas entre jogos que aconteceram e jogos que ainda vão acontecer

tabela_jogos_formatada["Resultado"] = tabela_jogos_formatada["Resultado"].fillna("a jogar") # add valor "a jogar" nas colunas sem valor
tabela_jogos_realizados = tabela_jogos_formatada[tabela_jogos_formatada["Resultado"].str.contains("–")] # pega as linhas com a coluna que possui resultados separados por "-"
display(tabela_jogos_realizados)
tabela_jogos_faltantes = tabela_jogos_formatada[~tabela_jogos_formatada["Resultado"].str.contains("–")] # "~" retorna o valor contrário 
tabela_jogos_faltantes = tabela_jogos_faltantes.drop(columns=["Resultado"]) # retira a coluna "resultaoo"
display(tabela_jogos_faltantes)
```

Para fazer a previsão, vamos utilizar os seguintes indicadores:
- média de gols dentro de casa
- média de gols fora de casa
- média de gols sofridos dentro de casa
- média de gols sofridos fora de casa

```python

# separando as colunas de gols fora e gols casa
tabela_jogos_realizados[["gols_casa", "gols_fora"]] = tabela_jogos_realizados["Resultado"].str.split("–", expand=True)
tabela_jogos_realizados = tabela_jogos_realizados.drop(columns=["Resultado"])
tabela_jogos_realizados["gols_casa"] = tabela_jogos_realizados["gols_casa"].astype(int)
tabela_jogos_realizados["gols_fora"] = tabela_jogos_realizados["gols_fora"].astype(int)
display(tabela_jogos_realizados)

media_gols_casa = tabela_jogos_realizados.groupby("Casa").mean(numeric_only=True) # transforma as colunas de gols em inteiros

# calculando média de gols dentro de casa
media_gols_casa = media_gols_casa.rename(columns={"gols_casa": "gols_feitos_casa", "gols_fora": "gols_sofridos_casa"})
display(media_gols_casa)

# calculando média de gols fora
media_gols_fora = tabela_jogos_realizados.groupby("Fora").mean(numeric_only=True)
media_gols_fora = media_gols_fora.rename(columns={"gols_casa": "gols_sofridos_fora", "gols_fora": "gols_feitos_fora"}) # renomeia as colunas
display(media_gols_fora)
```

```python
# mesclando as duas tabelas

tabela_estatisticas = media_gols_casa.merge(media_gols_fora, left_index=True, right_index=True) 
tabela_estatisticas = tabela_estatisticas.reset_index()
tabela_estatisticas = tabela_estatisticas.rename(columns={"Casa": "time"})
display(tabela_estatisticas)
```

# Distribuição de Poisson

Modelo estatístico usado para descrever a frequência com que um evento específico ocorre em um intervalo de tempo ou espaço, dado um conhecimento prévio da média de ocorrências desse evento.
- Parâmetro (lambda)
- a Distribuição de Poisson, o parâmetro λ e o Python podem ser explorados através da biblioteca scipy.stats do Python, que oferece funcionalidades para trabalhar com distribuições estatísticas, incluindo a Distribuição de Poisson
- em Python, podemos utilizar a distribuição de Poisson por meio da scipy.stats.poisson
- ideal para modelar diferentes situações onde a média de ocorrências de um evento é diferente

```python
from scipy.stats import poisson

# função para calcular a probabilidade do time ganhar/perder/empate partir das medias

def calcular_pontuacao_esperada(linha):
    time_casa = linha["Casa"]
    time_fora = linha["Fora"]

    # qual a média de gols que o time da casa faz contra o time de fora 
    lambda_casa = (tabela_estatisticas.loc[tabela_estatisticas["time"]==time_casa, "gols_feitos_casa"].iloc[0] 
                * tabela_estatisticas.loc[tabela_estatisticas["time"]==time_fora, "gols_sofridos_fora"].iloc[0])
    # média de gols time da casa * média de gols sofridos pelo time de fora - jogando fora


    # média de gols que o time de fora faz fora de casa * media de gols que o time da casa sofre dentro de casa
    lambda_fora =  (tabela_estatisticas.loc[tabela_estatisticas["time"]==time_fora, "gols_feitos_fora"].iloc[0] 
                    * tabela_estatisticas.loc[tabela_estatisticas["time"]==time_casa, "gols_sofridos_casa"].iloc[0])



# probabilidade de vitória em casa, probabilidade de empate, probabilidade de vitória fora de casa
    pv_casa = 0
    p_empate = 0
    pv_fora = 0

    for gols_casa in range(8):
        for gols_fora in range(8):
            probabilidade_resultado = poisson.pmf(gols_casa, lambda_casa) * poisson.pmf(gols_fora, lambda_fora)
            if gols_casa == gols_fora:
                p_empate += probabilidade_resultado
            elif gols_casa > gols_fora:
                pv_casa += probabilidade_resultado
            elif gols_casa < gols_fora:
                pv_fora += probabilidade_resultado

    # print(pv_casa)
    # print(p_empate)
    # print(pv_fora)

    ve_casa = pv_casa * 3 + p_empate
    ve_fora = pv_fora * 3 + p_empate
    linha["pontos_casa"] = ve_casa
    linha["pontos_fora"] = ve_fora
    return linha
```

```python
tabela_jogos_faltantes = tabela_jogos_faltantes.apply(calcular_pontuacao_esperada, axis=1)
display(tabela_jogos_faltantes)
```

```python
def ajustar_nome_time(linha):
    for nome in nomes_times:
        if nome in linha["Equipevde"]:
            return nome
    
tabela_classificacao["time"] = tabela_classificacao.apply(ajustar_nome_time, axis=1)
tabela_classificacao_atualizada = tabela_classificacao[["time", "Pts"]]
tabela_classificacao_atualizada["Pts"] = tabela_classificacao_atualizada["Pts"].astype(int)

display(tabela_classificacao_atualizada)
```

```python
tabela_pontuacao_casa = tabela_jogos_faltantes.groupby("Casa").sum(numeric_only=True)[["pontos_casa"]]
tabela_pontuacao_fora = tabela_jogos_faltantes.groupby("Fora").sum(numeric_only=True)[["pontos_fora"]]
display(tabela_pontuacao_casa)
display(tabela_pontuacao_fora)
```

```python
# atualizando tabela de acordo com o cálculo do valor esperado e ajustando os nomes das linhas

# agrupando e somando as colunas de pontuação 
def atualizar_pontuacao_previsao(linha):
    time = linha["time"]
    pontuacao = int(linha["Pts"]) + float(tabela_pontuacao_casa.loc[time, "pontos_casa"]) + float(tabela_pontuacao_fora.loc[time, "pontos_fora"])
    return pontuacao


tabela_classificacao_atualizada["Pts"] = tabela_classificacao_atualizada.apply(atualizar_pontuacao_previsao, axis=1)
tabela_classificacao_atualizada = tabela_classificacao_atualizada.sort_values(by="Pts", ascending=False) # atualiza as linhas (times) de acordo com a pontuação descrescente
tabela_classificacao_atualizada = tabela_classificacao_atualizada.reset_index(drop=True)
tabela_classificacao_atualizada.index = tabela_classificacao_atualizada.index + 1
display(tabela_classificacao_atualizada)
```
